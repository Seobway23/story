### 특정 추천 영화 조회
영화 검색-> 타이틀, 

TMDB에서 영화 검색, 
영화의 id값을 찾아 해당 영화에 대한 추천 영화 목록을 가져온다.

```python
import requests
from pprint import pprint

#무비 아이디를 찾아오는 법

title='그래비티'

URL = 'https://api.themoviedb.org/3/search/movie?'

params = {
    'api_key': '발급받아야 하는 키',
    'language': 'ko-KR',
    'region': 'KR',
    'query': title
}
response = requests.get(URL, params).json()
dict_response=response['results']

print(type(dict_response))

query_title_id=dict_response[0].get('id')
#가장 첫번째의 dict_response의 영화의 id를 할당하겠다!

print(query_title_id)



URL2 = f'https://api.themoviedb.org/3/movie/{query_title_id}/recommendations?'

params2 = {
    'api_key': '발급받아야 하는 키',
    'language': 'ko-KR',
    'region': 'KR',

}
response_2 = requests.get(URL2, params2).json()
dict_response_2=response_2['results']

print(dict_response_2)

dict_response_2_list=[]

for i in dict_response_2:
    #print(i)
    dict_response_2_list.append(i['title'])


print(dict_response_2_list)

```

```python
import requests
from pprint import pprint

#무비 아이디를 찾아오는 법

title='기생충'

URL = 'https://api.themoviedb.org/3/search/movie?'

params = {
    'api_key': '발급받아야 하는 키',
    'language': 'ko-KR',
    'region': 'KR',
    'query': title
}
response = requests.get(URL, params).json()
dict_response=response['results']

query_title_id=dict_response[0].get('id')
#가장 첫번째의 dict_response의 영화의 id를 할당하겠다!

print(query_title_id)

URL2 = f'https://api.themoviedb.org/3/movie/{query_title_id}/credits?'

params2 = {
    'api_key': '발급받아야 하는 키',
    'language': 'ko-KR',
    'region': 'KR',
}
response_2 = requests.get(URL2, params2).json()
dict_response_2=response_2['cast']
directing_response_2=response_2['crew']
#print(directing_response_2)


cast_list=[]
directing_list=[]

for idx in dict_response_2:
    if idx['cast_id']<10:
        cast_list.append(idx['name'])

cast_list_final={}
cast_list_final['cast']=cast_list

#print(directing_response_2)
for director in directing_response_2:
    if director['department']=='Directing':
        directing_list.append(director['name'])

directing_list_final={}
directing_list_final['directing']=directing_list

B=[cast_list_final,directing_list_final]
```
